```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data for the three models
metrics = {
    "Metric": ["MSE", "RMSE", "MAE", "R-squared", "Cross-Validated MSE", "Cross-Validated RMSE"],
    "Training_with_time": [0.004517, 0.067209, 0.031044, 0.964519, 0.016203, 0.127293],
    "Training_without_location": [0.014578, 0.120740, 0.067220, 0.885490, 0.136083, 0.368894],
    "Training_with_location": [0.004688, 0.068470, 0.032913, 0.963176, 0.018625, 0.136475]
}

# Plot each metric individually
metrics_df = pd.DataFrame(metrics).set_index("Metric")

for metric in metrics_df.index:
    plt.figure(figsize=(8, 2))
    plt.bar(
        metrics_df.columns,
        metrics_df.loc[metric],
        color=['blue', 'orange', 'green'],
        alpha=0.7
    )
    plt.title(f"{metric} Across Models")
    plt.ylabel(metric)
    plt.xlabel("Models")
    plt.tight_layout()
    plt.show()
```
